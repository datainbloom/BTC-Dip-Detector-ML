## Overview of Bitcoin**
**Bitcoin History**
Bitcoin is the worlds first cryptocurrency which is a type of digital money that runs on a decentralized network (called blockchain) instead of banks. It allows people to send and receive value directly, without intermediaries.

It was created in 2008 by an anonymous entity named Satoshi Nakamoto. The network launched in 2009. Over the years, it grew into one of the most talked-about assets in modern finance.

**Bitcoin Volatility**
Bitcoin is considered one of the most volatile assets because its price can swing dramatically in a very short time. Volatility means how much and how quickly the price of something goes up and down. How come this happens? Here are some reasons below:

- Low regulation - Bitcoin isnt tied to a company or government, so theres no central authority stabilizing its price unlike stocks or bonds.
- Supply & demand influence  Since Bitcoin has a limited supply, sudden surges or drops in demand have outsized effects.
- News/Media Hype - any announcements about regulation, technology, or big companies adopting/rejecting it can cause massive price swings.
- Market size - The Bitcoin market is still small compared to global stock or bond markets, so even modest trades can move the price a lot.

Thus, Bitcoins price is highly unpredictable and changes rapidly, which makes it exciting for traders but risky for investors. Plus, it makes an interesting case study so we decided to go with Bitcoin for our project.

## Machine Learning Model Used

**Random Forest Algorithm**
A Random Forest algorithm is a type of machine learning model that makes predictions by combining the results of many decision trees.

**What does a decision tree mean?**
A decision tree in a sense is asking a series of yes/no questions to make a decision (e.g., "Is the weather cold?", "Is it snowing?", and so on).

With the random forest model, it builds a lot of decision trees, each one slightly different because it looks at random samples of the data and random subsets of features.

**How does a Random Forest algorithm make predictions?**
For classification (e.g., "spam" or "not spam"), each tree votes, and the majority wins.
For regression (predicting a number), the trees predictions are averaged.

Because it uses many trees instead of one tree, a random forest is usually more accurate and less likely to make mistakes from overfitting. 

**What do you mean overfitting?**
Overfitting occurs when a model learns the training data very well (almost perfect), it learns all the tiny details and random noise instead of the general patterns. On the training data, it performs very well (almost perfect) on the training data. However, on new & unseen data, it performs poorly because it cant generalize.

For example, memorizing every question and answer for a practice test to ace the test. However the real exam has similar but different questions, you might struggle because you didnt learn the concepts, you only memorized the practice test.

**Why did we use the Random Forest algorithm?**
We chose the Random Forest algorithm because were specifying a select number of predictors to use. Also, it is a group of decision trees working together to make a stronger, more accurate prediction.
